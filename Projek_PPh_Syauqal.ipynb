## Data cleaning

### Subtask:
Clean the data to handle missing values and inconsistencies.


**Reasoning**:
Impute missing numerical values using the median, remove rows with missing values in crucial columns, and remove duplicates.



```python
# Impute missing numerical values using the median
numerical_cols = df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    if df[col].isnull().any():
        df[col].fillna(df[col].median(), inplace=True)

# Remove rows with missing values in crucial columns (e.g., 'tahun', 'pendapatan_operasional')
df.dropna(subset=['tahun', 'pendapatan_operasional', 'laba_sebelum_pajak'], inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Display descriptive statistics for numerical columns after cleaning
print(df.describe())

# Display the count of missing values after cleaning
print(df.isnull().sum())

display(df.head())
```
